````python
from datetime import datetime, timedelta
````

````python
start_date = datetime(2022, 1, 1)
end_date = datetime(2022, 12, 31)

def request_data_from_range(site_id: str, start_date: datetime, end_date:datetime, chunks=100):
    csv_files = []
    
    current_start_date = start_date

    while current_start_date < end_date:
        current_end = min(current_start_date + timedelta(days=chunks - 1), end_date)

        cur_start = fmt_date_to_nrel(current_start_date)
        cur_end = fmt_date_to_nrel(current_end)

        print(f"Fetching {site_id} date range {cur_start}-{cur_end}")
        code, text = wrangle_utils.request_station_data(site_id, cur_start, cur_end)

        file_path = f"data/site_data/bms/{site_id}-{cur_start}-{cur_end}.csv"

        with open(file_path, "w") as f:
            f.write(text)

        csv_files = [*csv_files, file_path]

        current_start_date = current_end + timedelta(days=1)

    return csv_files


def fmt_date_to_nrel(dt: datetime):
    return dt.strftime("%Y%m%d")
````

## Applying mappings and merging datasets

````python
import os
import pytz
````

````python
tz_map = {
    'MST': 'US/Mountain',
    'CST': 'US/Central',
    'HST': 'Pacific/Honolulu'
}
````

````python
def convert_to_utc(tz, time_col, row):
    # Pad time with zeros (e.g., 45 -> '0045')
    time_str = f"{int(row[time_col]):04d}"
    hour = int(time_str[:2])
    minute = int(time_str[2:])

    base_date = datetime(int(row['Year']), 1, 1) + timedelta(days=int(row['DOY']) - 1)
    local_dt = datetime(base_date.year, base_date.month, base_date.day, hour, minute)

    # Localize and convert to UTC
    localized = tz.localize(local_dt)
    return localized.astimezone(pytz.utc).timestamp()
````

````python
for site_id, mappings in fully_mapped.items():
    if site_id == "BMS": # We've got to load from multiple datasets
        continue

    print(list(mappings.values()))

    path = f"data/site_data/{site_id}.csv"
    file_size = os.path.getsize(path)

    if file_size < 1000:
        print(f"Skipping {path}. File is only {file_size}. Probably only contains error")
        continue

    print(f"Attempting to read {site_id}'s data from {path}")

    df = pd.read_csv(path)

    time_col = next((col for col in df.columns if col.endswith('ST')), None)

    if time_col is None or time_col not in tz_map:
        raise ValueError(f"Could not find valid time column or timezone mapping for {time_col}")

    local_tz = tz_map[time_col]
    tz = pytz.timezone(local_tz)  # Get the actual timezone object

    columns_to_keep = ['Year', 'DOY']
    if time_col:
        columns_to_keep.append(time_col)

    first_three = df[columns_to_keep]
    mapped_fields = df[list(mappings.values())].rename(columns={v: k for k, v in mappings.items()})

    result = pd.concat([first_three, mapped_fields], axis=1)


    result['timestamp'] = result.apply(lambda row: convert_to_utc(tz, time_col, row), axis=1)

    result['site_id'] = site_id

    columns_order = ['site_id', 'timestamp'] + [col for col in result.columns if col not in {'site_id', 'timestamp'}]
    result = result[columns_order]

    result.drop(columns=columns_to_keep, errors='ignore', inplace=True)

    result.to_csv(f'data/site_data_cleaned/{site_id}.csv', index=False)
````

````python
bms_df = pd.DataFrame()
mappings = fully_mapped["BMS"]
site_id = "BMS"

for file in bms_files:
    path = file
    file_size = os.path.getsize(path)

    if file_size < 1000:
        print(f"Skipping {path}. File is only {file_size}. Probably only contains error")
        continue

    print(f"Attempting to read {site_id}'s data from {path}")

    df = pd.read_csv(path)

    time_col = next((col for col in df.columns if col.endswith('ST')), None)

    if time_col is None or time_col not in tz_map:
        raise ValueError(f"Could not find valid time column or timezone mapping for {time_col}")

    local_tz = tz_map[time_col]
    tz = pytz.timezone(local_tz)  # Get the actual timezone object

    columns_to_keep = ['Year', 'DOY']
    if time_col:
        columns_to_keep.append(time_col)

    first_three = df[columns_to_keep]
    mapped_fields = df[list(mappings.values())].rename(columns={v: k for k, v in mappings.items()})

    result = pd.concat([first_three, mapped_fields], axis=1)


    result['timestamp'] = result.apply(lambda row: convert_to_utc(tz, time_col, row), axis=1)

    result['site_id'] = site_id

    columns_order = ['site_id', 'timestamp'] + [col for col in result.columns if col not in {'site_id', 'timestamp'}]
    result = result[columns_order]

    result.drop(columns=columns_to_keep, errors='ignore', inplace=True)

    bms_df = pd.concat([bms_df, result], ignore_index=True)
````
